```python
class Shape: 
    def identity(self,sides):
        if sides==0:
         print("circle")
        elif sides==3:
         print("triangle")
        elif sides==4:
         print("square")
        else:
         print("invalid sides")
obj=Shape()
obj.identity(0)
obj.identity(3)
obj.identity(4)
```

```python
#nested classes or inner classes
class Employee:
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal
class Employee_2:
    def __init__(self,name,sal):
        self.cname=name
        self.csal=sal
s=Employee("deepu","50,000")
s1=Employee_2("neelu","1,00,000")
print(s.name,s.sal,s1.cname,s1.csal)
```

```python
#ANOTHER EXAMPLE-INNER CLASS
class Laptop:
    def __init__(self, type, storage):
        self.type = type
        self.storage = storage

    class battery:
        def __init__(self, version):
            self.version = version

# create objects
a = Laptop("64bit", "232gb")
b = Laptop.battery("H32")

# print output
print(a.type, a.storage, b.version)
```

```python
#inheritance
#single
class A:
    def view(self):
        print("this is base class")
class B(A):
    def view1(self):
        print("this is derived class")
obj=B()
obj.view()
obj.view1()
```

```python
#multi-level inheritance -like chain(child gets properties from both parent and grandparents)
class A:
    def view(self):
        print("this is base class")
class B(A):
    def view1(self):
        print("this is derived class")
class C(B):
    def view2(self):
        print("this is another derived class")
obj=C()
obj.view()
obj.view1()
obj.view2()
```

```python
#hierarichal-many child class inherits from one parent class
class A:
    def view(self):
        print("this is base class")
class B(A):
    def view1(self):
        print("this is derived class")
class C(A):
    def view2(self):
        print("this is another derived class")
obj1=B()
obj.view()
obj.view1()
obj2=C()
obj.view()
obj.view2()
```

```python
#multiple inheritance-many parent classes inherits one child
class A:
    def display(self):
        print("this is base class")
class B:
    def display1(self):
        print("this is second base class")
class C:
    def display2(self):
        print("this is third base calss")
class D(A,B):
    def display3(self):
        print("this is child class")
obj=D()
obj.display3()
obj.display()
obj.display1()
obj1=C()
obj1.display2()
```

```python
#hybrid inheritance
class A:
    def view(self):
        print("this is base class")
class B(A):
    def view1(self):
        print("this is derived class")
class C(A):
    def view2(self):
        print("this is another derived class")
class D(B,C):
    def display(self):
        print("this is last derived class")
obj=D()
obj.display()
obj.view()
obj.view1()
obj.view2()
```

```python
#method overriding-same methods in both classes but different behavior
class Vehicle:
    def sound(self):
        print("this is vehicle")

class Bike(Vehicle):
    def sound(self):
        super().sound() 
        print("this is bike")
obj = Bike()
obj.sound()
```
