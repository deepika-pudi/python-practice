```python
#the string palindrome or not
str="madam"
str1=str.reverse("madam")
if(str1==str):
    print("string is a palindrome")
else:
    print("string is not a palindrome")
```

```python
#the string palindrome or not
str="madam"
str1=str[::-1]
if(str==str1):
    print("string is a palindrome")
else:
    print("string is not a palindrome")
```
