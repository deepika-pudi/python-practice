```python
#set operayions 
#intersection-common elemnets
s={1,4,2,6,7}
s1={4,6,7,5,8}
s3=s&s1
print(s3)
```

```python
#union-all elements in set(|)
s={1,4,2,6,7}
s1={4,6,7,5,8}
s3=s|s1
print(s3)
```

```python
#set difference-finding element in one set which is not in another set
s={1,4,2,6,7}
s1={4,6,7,5,8}
s3=s-s1 #  if s-s1 result is s ,else s1-s result is s1 
print(s3)
```
