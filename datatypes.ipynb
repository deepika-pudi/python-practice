```python
#code on data types
# numeric->int
a=100
print("integer:", a, type(a))
```

```python
#float
a=5.2
print("float:", a, type(a))
```

```python
#complex
d=2+4j
print("complex:", d, type(d))
```

```python
#sequence->string
name="deepu"
print("string:", name, type(name))
```

```python
#list(ordered,changable)
a=['drink','coffee','tea']
print("list:", a, type(a))
```

```python
fruits = ["apple", "banana", "mango"]
print("List:", fruits, type(fruits))
```

```python
#tuple(unordered, not changed)
a=(1,2,3,4)
print("tuple:", a, type(a))
```

```python
#mapping->dict
employee = {
    "name":"krishna",
    "id" : 2045,
    "dept": 12,
}
print("keyvaluepairs:", employee, type(employee))
```

```python
#set->set(unordered, no duplicate values)
a={1,2,3,4}
print("set:", a, type(a))
```

```python
#boolean
a=True
print("bool:", a , type(a))
```

```python
#swapping of numbers in different ways
#using third variable
a=10
b=20
temp=a
a=b
b=temp
print(a,b)
```

```python
#without using third variable
a=10
b=20
a,b=b,a
print(a,b)
```

```python
#using addition and subtraction(works for numbers)
a=2
b=5
a=a+b
b=a-b
a=a-b
print(a,b)
```

```python
#using mul and div(doesnot work if one number 0)
a=20
b=40
a=a*b
b=a/b
a=a/b
print(int(a),int(b))
```

```python
#using xor
a = 10
b = 20

a = a ^ b
b = a ^ b
a = a ^ b

print(a, b)
```
