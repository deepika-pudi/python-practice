```python
#types of arguments
#positional-values are passed as same order as parameters passed
#keyword-values are passed in agruments using their parameter values in any order
#default-parameters that have a default value. If no value is given when calling the function, Python uses the default value.
```

```python
#positional
def greet(name,age):
    print(f"hello {name}, you are {age} years old")
greet("john","30")
```

```python
#keyword
def greet(name,age):
 print(f"hello {name}, you are {age} years old")
greet(age=30,name="virat")
```

```python
#default parameters
def greet(name="deepu"):
    return name
print(greet())
```

```python
#work-correct answers only
def prime(n):
    for i in range(2, n):
        if n % i == 0:
            print("Not a prime number")
            return
    print("Prime number")
```

```python
def add(a, b):
    print("Sum =", a + b)

def sub(a, b):
    print("Difference =", a - b)

def mul(a, b):
    print("Product =", a * b)

def div(a, b):
    print("Division =", a / b)

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

add(a, b)
sub(a, b)
mul(a, b)
div(a, b)
```

```python
#*agrs-used when no of parameters are unknown
#variable-length arguments
def add(*args):
    total=sum(args)
    return total
a=add(3,5,6)
print(a)
```

```python
#variable-length keyword arguments
#**kwargs-used to store arguments with parameter values
def greet(**kwargs):
 for key,value in kwargs.items():
    print(f"{key}:{value}")
greet(name="deepu",age=20)
```

```python
#sample program
def subtract(*args):
    total=args[0]-sum(args[1:])
    return total
a=subtract(10,4,3)#for 3 parameters
print(a)
```

```python
def sub(*args):
    total = args[0] - args[1]
    return total

a = sub(10, 4)#for two parameters
print(a)
```

```python
def mul(*args):
    total = args[0] * args[1] * args[2]
    return total

a = mul(2, 3, 4)
print(a)
```

```python
def div(*args):
    total = args[0] / args[1] / args[2]
    return total

a = div(20, 2, 5)
print(a)
```

```python
#calculator
def cal(a,b,op):
    if op=='+':
        return a+b
    elif op=='-':
        return a-b
    elif op=='*':
        return a*b
    elif op=='/':
       if b==0:
           print("invalid")
       else:
        print("a/b")
    else:
        print("invalid section")
a=int(input("enter a value"))
b=int(input("enter b value"))
op=input("select any operation '+','-','*','/':")
cal(a,b,op)
```
