```python
'''lambda functions
syntax:lambda arguments:expression'''
#square 
sq=lambda v : v**2
sq(7)
```

```python
#addition
add=lambda a,b : a+b
add(5,5)
```

```python
#using map()-function appplies each element in the list
#square
li=[1,2,3,4,5]
n=list(map(lambda m : m**2,li))
print(n)
```

```python
#using filter()-after satisfing the condition it return a value
#even numbers
l1=[1,2,3,5,6,7,8]
m=list(filter(lambda v : v%2==0,l1))
print(m)
```

```python
#using reduce()-convert multiple data into single value
#addition
from functools import reduce
r=[1,2,3,4]
s=reduce(lambda a,b : a+b,r)
print(s)
```

```python
#work
l=[1,2]
s=[]
for i in l:
    s.append(i*2)
print(s)
```

```python
#using map
l=[1,2]
s=list(map(lambda n : n*2,l))
print(s)
```

```python
l=[2,4]
n=(lambda a,b : a if a>b else b)(l[0],l[1])
print(n)
```
