```python
mar1,mar2,mar3= map(int,input("enter marks for std1, std2,std3").split())
if mar1>=90:
    print("A grade")
elif mar2>=80:
    print("B grade")
elif mar3>=70:
    print("c grade")
else:
    print("fail")
```

```python
marks=int(input("enter student marks(1-100):"))
if marks<0 or marks>100:
    print("invalid input! number must be b/w 1-100")
else:
   if marks>=90:
    print("A grade")
   elif marks>=80:
    print("B grade")
   elif marks>=70:
    print("C grade:")
   else:
    print("fail")
```

```python
#using if else condition even or not
num=int(input("enter any  number"))
if num%2 != 0:
    print("odd number")
else:
    print("even number")
```

```python
#checking a num is pos,neg or zero
num = int(input("enter a number"))
if num>0:
    print("positive number")
elif num<0:
    print("negative number")
else:
    print("whole number")
```

```python
#transfer statements-break
for i in range(5):
    if i==3:
        break
    print(i)
```
